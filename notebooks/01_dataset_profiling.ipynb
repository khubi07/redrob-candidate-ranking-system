```python
import json
import numpy as np
from tqdm import tqdm

skill_counts = []
career_counts = []
education_counts = []
cert_counts = []
language_counts = []

with open(DATA_PATH, "r", encoding="utf-8") as f:
    for line in tqdm(f):
        candidate = json.loads(line)

        skill_counts.append(len(candidate.get("skills", [])))
        career_counts.append(len(candidate.get("career_history", [])))
        education_counts.append(len(candidate.get("education", [])))
        cert_counts.append(len(candidate.get("certifications", [])))
        language_counts.append(len(candidate.get("languages", [])))
```

```python
def summarize(name, values):
    print(f"\n{name}")
    print("-" * 30)
    print(f"Average : {np.mean(values):.2f}")
    print(f"Median  : {np.median(values):.2f}")
    print(f"Minimum : {np.min(values)}")
    print(f"Maximum : {np.max(values)}")

summarize("Skills", skill_counts)
summarize("Career History", career_counts)
summarize("Education", education_counts)
summarize("Certifications", cert_counts)
summarize("Languages", language_counts)
```

```python
import json
import pandas as pd
from tqdm import tqdm

signal_data = []

with open(DATA_PATH, "r", encoding="utf-8") as f:
    for line in tqdm(f):
        candidate = json.loads(line)

        s = candidate["redrob_signals"]

        signal_data.append({
            "profile_completeness_score": s["profile_completeness_score"],
            "open_to_work_flag": s["open_to_work_flag"],
            "profile_views_received_30d": s["profile_views_received_30d"],
            "applications_submitted_30d": s["applications_submitted_30d"],
            "recruiter_response_rate": s["recruiter_response_rate"],
            "avg_response_time_hours": s["avg_response_time_hours"],
            "connection_count": s["connection_count"],
            "endorsements_received": s["endorsements_received"],
            "notice_period_days": s["notice_period_days"],
            "willing_to_relocate": s["willing_to_relocate"],
            "github_activity_score": s["github_activity_score"],
            "search_appearance_30d": s["search_appearance_30d"],
            "saved_by_recruiters_30d": s["saved_by_recruiters_30d"],
            "interview_completion_rate": s["interview_completion_rate"],
            "offer_acceptance_rate": s["offer_acceptance_rate"],
            "verified_email": s["verified_email"],
            "verified_phone": s["verified_phone"],
            "linkedin_connected": s["linkedin_connected"]
        })

signals_df = pd.DataFrame(signal_data)
signals_df.describe(include="all")
```

```python
from collections import Counter

skill_counter = Counter()

with open(DATA_PATH, "r", encoding="utf-8") as f:
    for line in f:
        candidate = json.loads(line)

        for skill in candidate.get("skills", []):
            skill_counter[skill["name"]] += 1

skill_counter.most_common(30)
```

```python
import json
import random

sample_descriptions = []

with open(DATA_PATH, "r", encoding="utf-8") as f:
    for i, line in enumerate(f):
        candidate = json.loads(line)

        for exp in candidate.get("career_history", []):
            desc = exp.get("description")

            if desc and len(desc) > 50:
                sample_descriptions.append(desc)

print("Collected:", len(sample_descriptions))
```

```python
for desc in random.sample(sample_descriptions, 10):
    print("=" * 80)
    print(desc)
    print()
```

```python
keywords = [
    "retrieval",
    "search",
    "ranking",
    "recommendation",
    "embedding",
    "relevance",
    "machine learning",
    "ml",
    "ai",
    "llm"
]

matches = {k: 0 for k in keywords}

with open(DATA_PATH, "r", encoding="utf-8") as f:
    for line in f:
        candidate = json.loads(line)

        for exp in candidate.get("career_history", []):
            desc = exp.get("description", "").lower()

            for k in keywords:
                if k in desc:
                    matches[k] += 1

matches
```
